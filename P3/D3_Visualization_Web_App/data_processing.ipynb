```python
import pandas as pd
import json;
from pprint import pprint
```

```python
df = pd.read_csv("../quarterly_greenhouse.csv")
value_cols = [c for c in df.columns if c.endswith(("Q1","Q2","Q3","Q4"))]

long = (
    df.melt(id_vars=[c for c in df.columns if c not in value_cols],
            value_vars=value_cols,
            var_name="quarter",
            value_name="emissions")
      .dropna(subset=["emissions"])
)

quarter_to_date = {"Q1": "-03-31", "Q2": "-06-30", "Q3": "-09-30", "Q4": "-12-31"}
long["date"] = pd.to_datetime(
    long["quarter"].str[:4] + long["quarter"].str[4:].map(quarter_to_date)
)

long.to_csv("../quarterly_greenhouse_long.csv", index=False)
long.to_json("../quarterly_greenhouse_long.json", orient="records", date_format="iso")
```
